```python
import time
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

import scipy as sp
import scipy.sparse.linalg as linalg
import scipy.cluster.hierarchy as hr
import sklearn.datasets as datasets
import sklearn.metrics as metrics
import sklearn.linear_model as linear_model
import statsmodels.api as sm
from patsy import dmatrices

import seaborn as sns

import scrape_etsy as scrape
```

```python
results2 = scrape.hats(240)
```

```python
%matplotlib inline
df = pd.DataFrame(results2)
df = df.T
df.Location = df.Location.str.replace("-1", '')
df = df.convert_objects(convert_numeric=True)
averages=[]
ranges=[]
for i in range(1,len(df)+1):
    averages.append(np.mean(df.loc[i].Prices))
    ranges.append(max(df.loc[i].Prices) - min(df.loc[i].Prices))
df['Avg_price'] = averages
df['Rank'] = df.index
df['Price_range'] = ranges
df['Approx_daily_profit'] = df.Avg_price * df.Sales / df.Age
df
```

```python
filtered = df[df.Avg_price>1]
X = filtered.Approx_daily_profit
Y = filtered.Rank
plt.scatter(X, Y, s=3, color='orange')
plt.xlabel('Approximate Daily Profit')
plt.ylabel('Rank')

#Linear Regression for Rank
r = filtered.copy()
r['rand'] = [np.random.rand() for i in range(len(r))]
r = r.sort('rand')
r = r.drop('rand', 1)
r = r.convert_objects(convert_numeric=True)

test = r.head(len(r)/2).replace(-1, 0)
train = r.tail(len(r)/2).replace(-1,0)

y_train = train[['Rank']]
y_test = test[['Rank']]

x_train = train[['Approx_daily_profit']]
x_test = test[['Approx_daily_profit']]

#LINEAR REGRESSION
regr = linear_model.LinearRegression()
regr.fit(x_train, y_train)

print("Training error: ", metrics.mean_squared_error(regr.predict(x_train),y_train))
print("Test     error: ", metrics.mean_squared_error(regr.predict(x_test),y_test))

train_score = regr.score(x_train,y_train)
test_score = regr.score(x_test,y_test)
print("Training score: ", train_score)
print("Test     score: ", test_score)

features = ['Approx_daily_profit']
coefficients = regr.coef_
for i in range(len(coefficients)):
    print features[i],"\t",coefficients[i]
    
#OLS REGRESSION
model = sm.OLS(y_train, x_train)
results = model.fit()
print results.summary()
print "Confidence Intervals:", results.conf_int()
print "Parameters:", results.params

plt.plot(x_test, regr.predict(x_test), color='red',
         linewidth=1)
```
