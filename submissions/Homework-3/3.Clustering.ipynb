## Step 1. Simple clustering

The first thing that we will try is to cluster the geolocation data of the calls. The geolocation of a record consists of a pair `(latitude, longitude)`, which represents the position of the record on the map. Since the dataset that we have spans only a small geographic location, i.e. Boston, we will use the Euclidean distance when measuring how close two points are. Techinically, we should pick a [projection](https://github.com/mbostock/d3/wiki/Geo-Projections) in order to transform the data into 2d, and measure  the Euclidean distance in the projected space.

Since we are dealing with numeric data, we might consider centering and normalizing them. In order to do that, check the [StandardScaler](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler) from `scikit-learn`.

For exampe, the result of a clustering with 10 clusters is the following:


![A clustering of the locations of the calls for k=10](clusters10.png)

Filter the dataset to get the latitude and longitude, transform it as required and run the k-means clustering algorithm. Argue about the number of clusters you used. **(pts 50)**

```python
locs = boston.Geocoded_Location
#place coordinate data into arrays
a=[]
x=[]
y=[]
for entry in locs:
    split = entry.split(',')
    X = split[0][1:]
    Y = split[1][:-1]
    a.append([X, Y])
    x.append(X)
    y.append(Y)
a = np.array(a)
x = np.array(x)
y = np.array(y)

#transform data
scaler = sk.preprocessing.StandardScaler()
x = scaler.fit_transform(x)
y = scaler.fit_transform(y)

l = np.array([x,y])
l = np.swapaxes(l, 0, 1)

#find kmeans clusters
kmeans = KMeans(init='k-means++', n_clusters=6, n_init=10)
kmeans.fit_predict(l)
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
error = kmeans.inertia_

#calculate ideal number of clusters
def evaluate_clusters(X,max_clusters):
    error = np.zeros(max_clusters+1)
    error[0] = 0;
    for k in range(1,max_clusters+1):
        kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
        kmeans.fit_predict(X)
        error[k] = kmeans.inertia_

    plt.plot(range(1,len(error)),error[1:])
    plt.xlabel('Number of clusters')
    plt.ylabel('Error')

evaluate_clusters(l, 20)

#plot
fig, a = plt.subplots(1,1,figsize=(10,10))
cmap = sns.color_palette("bright", n_colors=4)
cmap.extend(sns.hls_palette(n_colors=6, h=0.7, l=0.4, s=0.8))

for i, label in enumerate(set(labels)):
    point_indices = np.where(labels == label)[0]
    point_indices = point_indices.tolist()
    plt.scatter(y[point_indices], x[point_indices], s=25, alpha=0.2, c=cmap[i], marker='+')
```

The total number of clusters used was 6. According to the graph generated by evaluate_clusters(), the biggest decrease
in error occurs around 5 clusters. By looking at the scatterplot, there are two clear groups of location data that are 
distanced from the rest, one in the upper-left corner and one in the upper-right corner. A scatterplot with 5 clusters
doesn't accurately separate these two groups. A scatterplot with 6 clusters divides the data in a more helpful way.

## Step 2. Using multiple features

The next step in our analysis is to use more features than just the geolocation of a record (whose clustering only gives us information about the neighborhoods, which should not be a huge revelation!) Filter the dataset to get more features, transform them as required and compute another clustering. You should argue about the quality of your clustering and use domain knowledge (i.e. the fact that you are locals) to explore the clusters you produced and present them to the reader. **(pts 30)**

```python
reasondata = boston.REASON
datesdata = boston.OPEN_DT
reasons=[]
dates=[]

#assign values to each reason, and store data in an array
reasons_dict = {}
count = 0
for x in range(len(reasondata)):
    if reasondata[x] in reasons_dict:
        continue
    else: 
        reasons_dict[reasondata[x]] = count
        count = count + 1
        
for x in range(len(datesdata)):
    if reasondata[x] == reasondata[x]:
        reasons.append(reasons_dict[reasondata[x]])
    else:
        reasons.append(-1)

#store dates in an array
for x in range(len(datesdata)):
    value = datesdata[x][5:7]+datesdata[x][8:10]
    dates.append(value)

#transform data
scaler = sk.preprocessing.StandardScaler()
dates = scaler.fit_transform(dates)
reasons = scaler.fit_transform(reasons)

T = np.array([dates,reasons])
T = np.swapaxes(T, 0, 1)

#calculate ideal number of clusters
def evaluate_clusters(X,max_clusters):
    error = np.zeros(max_clusters+1)
    error[0] = 0;
    for k in range(1,max_clusters+1):
        kmeans = KMeans(init='k-means++', n_clusters=k, n_init=10)
        kmeans.fit_predict(X)
        error[k] = kmeans.inertia_

    plt.plot(range(1,len(error)),error[1:])
    plt.xlabel('Number of clusters')
    plt.ylabel('Error')

evaluate_clusters(T, 20)

#find kmeans clusters
kmeans = KMeans(init='k-means++', n_clusters=5, n_init=10)
kmeans.fit_predict(T)
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
error = kmeans.inertia_

#plot
fig, a = plt.subplots(1,1,figsize=(10,10))
cmap = sns.color_palette("bright", n_colors=4)
cmap.extend(sns.hls_palette(n_colors=6, h=0.7, l=0.4, s=0.8))

for i, label in enumerate(set(labels)):
    point_indices = np.where(labels == label)[0]
    point_indices = point_indices.tolist()
    plt.scatter(dates[point_indices], reasons[point_indices], s=25, alpha=0.2, c=cmap[i], marker='+')
    plt.xlabel('Dates')
    plt.ylabel('Reasons')
```

When looking at the complaint reasons by day, you can visualize how the range of reasons for complaints has expanded throughout the winter. The red cluster at the top was barely visible near the beginning, but in later dates it became more dense. This shows that new categories for complaint reasons have appeared in late winter, suggesting that the bad weather has created many types of problems that didn't exist before.

## Step 3. Detecting outliers

Now that we have a clustering of the records in the dataset, we can detect the outlying ones. These are the ones who are the farthest from the centroids of their clusters. Track them down and describe any interesting observations that you can make. **(pts 20)** 

```python
#calculate kmeans
kmeans = KMeans(init='k-means++', n_clusters=6, n_init=10)
kmeans.fit_predict(l)
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
error = kmeans.inertia_

#finds the euclidean distance between two points
def euclidean_distance(point1, point2):
    x1 = point1[0]
    y1 = point1[1]
    x2 = point2[0]
    y2 = point2[1]
    return np.sqrt(np.square(x2-x1)+np.square(y2-y1))

#create an array containing the distance between each point and its centroid
distances=[]
for x in range(len(l)):
    distances.append(euclidean_distance(l[x], centroids[labels[x]]))

#calculate the average distance from the center of each cluster to all of its points
sum0=0
sum1=0
sum2=0
sum3=0
sum4=0
sum5=0
count0=0
count1=0
count2=0
count3=0
count4=0
count5=0
for x in range(len(l)):
    if labels[x]==0:
        sum0=sum0+distances[x]
        count0=count0+1
    elif labels[x]==1:
        sum1=sum1+distances[x]
        count1=count1+1
    elif labels[x]==2:
        sum2=sum2+distances[x]
        count2=count2+1
    elif labels[x]==3:
        sum3=sum3+distances[x]
        count3=count3+1
    elif labels[x]==4:
        sum4=sum4+distances[x]
        count4=count4+1
    elif labels[x]==5:
        sum5=sum5+distances[x]
        count5=count5+1
avg0=sum0/count0
avg1=sum1/count1
avg2=sum2/count2
avg3=sum3/count3
avg4=sum4/count4
avg5=sum5/count5

#target outliers if their distance is over 2 times greater than average
outliers=[]
for x in range(len(l)):
    if labels[x]==0:
        if distances[x]> 2*avg0:
            outliers.append(l[x])
    if labels[x]==1:
        if distances[x]> 2*avg1:
            outliers.append(l[x])
    if labels[x]==2:
        if distances[x]> 2*avg2:
            outliers.append(l[x])
    if labels[x]==3:
        if distances[x]> 2*avg3:
            outliers.append(l[x])
    if labels[x]==4:
        if distances[x]> 2*avg4:
            outliers.append(l[x])
    if labels[x]==5:
        if distances[x]> 2*avg5:
            outliers.append(l[x])
outliersx=[]
outliersy=[]
for x in outliers:
    outliersx.append(x[0])
for x in outliers:
    outliersy.append(x[1])
plt.plot(outliersx, outliersy, 'bo')

```

The outliers shown above are all more than 2 times farther from their cluster centers than average. They are all locations on the outskirts of the original location data.
